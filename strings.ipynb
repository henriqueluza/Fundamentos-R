## Trabalhando com Strings

### `paste()` - Juntar Textos

```r
paste("Olá", "Mundo")
# Resultado: "Olá Mundo"
```

### `nchar()` - Contar caracteres em uma string

```r
nchar("Brasil")
# Resultado: 6
```

### `toupper()` e `tolower()`- Maiúsculo e minúsculo

```r
toupper("brasil")
# Resultado: "BRASIL"

tolower("BRASIL")
# Resultado: "brasil"
```

###  `substr()` - Extrair parte do texto

```r
# substr(texto, início, fim)
substr("COVID-19", 1, 5)
# Resultado: "COVID"

substr("17 Dec 2020", 4, 6)
# Resultado: "Dec"
```

### `strsplit()` - Dividir texto

```r
strsplit("Brasil,USA,India", split = ",")
# Resultado: list("Brasil", "USA", "India")

# Para pegar como vetor:
unlist(strsplit("Brasil,USA,India", split = ","))
# Resultado: c("Brasil", "USA", "India")
```

### `gsub()` e `sub()` - Substituir texto

```r
# Remover vírgulas de números
gsub(",", "", "1,234,567")
# Resultado: "1234567"

# Substituir texto
gsub("Covid", "COVID", "Covid-19 é uma doença")
# Resultado: "COVID-19 é uma doença"

# Remover espaços extras
gsub(" +", " ", "Brasil    USA")
# Resultado: "Brasil USA"

sub(",", "", "1,234,567")
# Resultado: "1234567"  (só remove a primeira vírgula!)
```

### `grep()` e `grepl()` - Buscar padrões

```r
# Verificar se contém o texto
grepl("Bra", "Brasil")
# Resultado: TRUE

grepl("USA", "Brasil")
# Resultado: FALSE

# Útil para filtrar dados! 
covid_data %>% filter(grepl("Bra", Country))

paises <- c("Brasil", "USA", "Brunei", "India")

grep("Bra", paises)
# Resultado: 1  (posição do "Brasil")

grep("Br", paises)
# Resultado: c(1, 3)  (Brasil e Brunei)

# Para retornar os valores:
grep("Br", paises, value = TRUE)
# Resultado: c("Brasil", "Brunei")
```

### `trimws()` - Remover espaços em branco

```r
trimws("  Brasil  ")
# Resultado: "Brasil"

trimws("  Brasil  ", which = "left")
# Resultado: "Brasil  "

trimws("  Brasil  ", which = "right")
# Resultado: "  Brasil"
```
