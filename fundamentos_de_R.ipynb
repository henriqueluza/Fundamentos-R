<h3>Data Structures</h3>

```r
my_vector <- c(100, 200, 400) #Vetores
print(my_vector * 2)
print(my_vector[1])

my_list <- list(
  name = "Henrique",
  age = 21,
  salary = 1300
)

print(my_list["age"])

my_matrix <- matrix(
  c(1, 0, 0, 1),
  nrow = 2,
  ncol = 2
)

print(my_matrix)
print(my_matrix[1, 1]) # linha e coluna, nessa ordem

my_matrix2 <- matrix(
  c(5, 6, 7, 6),
  nrow = 2,
  ncol = 2
)

sumn_of_matrices <- my_matrix + my_matrix2
print(sumn_of_matrices)

print(my_matrix * sumn_of_matrices)

transposed_matrix <- t(my_matrix2)
print(transposed_matrix)

shape_of_matrix <- dim(transposed_matrix)
print(shape_of_matrix)

my_data_frame <- data.frame(
  name = c("Henrique", "Edna", "Claudinei"),
  age = c(21, 55, 57),
  job = c("Estudante", "Professora", "Servidor Público")
)

print(my_data_frame)
print(my_data_frame$name)
print(my_data_frame[2, "age"])
```

<h3>Control Flow</h3>

```r
var <- 0

if (var / 2 == 0) {
  print("Var is even")
} else if (var / 2 == 1) {
  print("Var is odd")
}

for (patient_id in 1001:1005) {
  print(paste("Processing patient", patient_id))
}

month <- 1
total_expenses <- 0

while (total_expenses < 5000) {
  total_expenses <- total_expenses + 600  # monthly expense
  print(paste("Month", month, "- Total:", total_expenses))
  month <- month + 1
}

repeat {
  count_errors <- sample(0:5, 1)  # simulate checking a random error count
  print(paste("Errors found:", count_errors))
  if (count_errors == 0) {
    print("No errors found—stopping check.")
    break
  }
}

for (i in 1:5) {
  if (i == 3) {
    next  # skips the current iteration
  }
  print(i)
}
```

<h3>Functions</h3>

```r
calculate_total <- function(amount, tax_rate = 0.10) {
  # especifica valor padrão para a tax_rate
  total <- amount + (amount * tax_rate)
  return(total)
}

total1 <- calculate_total(100)        # uses default tax_rate 10%
total2 <- calculate_total(100, 0.15)  # uses specified tax_rate 15%

print(total1)  # Output: 110
print(total2)  # Output: 115

# Lembrar de return nas funções e lembrar de descrever elas com comentários
```

```r
readings <- c(52, 48, NA, 56)

if (any(is.na(readings))) {
  print("Warning: Data contains missing values.")
}


amount_paid <- "twenty"

if (!is.numeric(amount_paid)) {
  print("Error: Amount must be a number.")
}


validate_ridership <- function(data) {
  if (!is.numeric(data)) {
    return("Ridership data must be numeric.")
  }
  if (any(is.na(data))) {
    return("Ridership data contains missing values.")
  }
  "Ridership data is ready to use."
}
```
